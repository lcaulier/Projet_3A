```python
n=0 # nombre d'images
i=0 # compteur pour parcourir le fichier
filename = "produit.jpg"
foldername_front = "images_front"
foldername_nutrition = "images_nutrition"
foldername_ingredients = "images_ingredients"
if not os.path.exists(foldername):
    os.mkdir(foldername)
while n <= 5:
  #random = randint(i,i+200)
  #print(random)
  data = pd.read_csv("en.openfoodfacts.org.products.csv\en.openfoodfacts.org.products.csv", sep='\t', skiprows=i,nrows=1)
  prod_id=data.iloc[0,0]
  liste_urls= [data.iloc[0,79],data.iloc[0,81],data.iloc[0,83]]
  if np.NaN not in liste_urls:
    try:
      rep = requests.get(liste_urls[0])
      mpimg.imsave("front_"+prod_id+".png", rep)
      rep = requests.get(liste_urls[1])
      mpimg.imsave("nutrition_"+prod_id+".png", rep)
      rep = requests.get(liste_urls[2])
      mpimg.imsave("ingredients_"+prod_id+".png", rep)
      n+=1
    except:
      print("Impossible de trouver les images pour la ligne",i)
  i+=1
```

```python
foldername_front = "Dataset_couleur/images_front"
foldername_nutrition = "Dataset_couleur/images_nutrition"
foldername_ingredients = "Dataset_couleur/images_ingredients"
if not os.path.exists(foldername_front):
    os.mkdir(foldername_front)
if not os.path.exists(foldername_ingredients):
    os.mkdir(foldername_ingredients)
if not os.path.exists(foldername_nutrition):
    os.mkdir(foldername_nutrition)
```

```python
i=0 # counter to parse the file
step = 100 # step to the next rows
x=0 # number of front images
y=0 # number of ingredients images
z=0 # number of nutritional images
limit = 2000 # If we want 2000 images per categories

while x <limit or y<limit or z<limit:
  random = randint(i,i+100) 
  data = pd.read_csv("en.openfoodfacts.org.products.csv\en.openfoodfacts.org.products.csv", sep='\t', skiprows=random,nrows=1)
  prod_id=data.iloc[0,0]
  if not pd.isna(data.iloc[0,79]) and x<limit:
    filepath = os.path.join(foldername_front, "front_"+str(prod_id)+".jpg")
    rep = requests.get(data.iloc[0,79])
    with open(filepath, "wb") as f:
        f.write(rep.content)
    x+=1
  if not pd.isna(data.iloc[0,81]) and y<limit:
    filepath = os.path.join(foldername_ingredients, "ingredients_"+str(prod_id)+".jpg")
    rep = requests.get(data.iloc[0,81])
    with open(filepath, "wb") as f:
        f.write(rep.content)
    y+=1
  if not pd.isna(data.iloc[0,83]) and z<limit:
    filepath = os.path.join(foldername_nutrition, "nutrition_"+str(prod_id)+".jpg")
    rep = requests.get(data.iloc[0,83])
    with open(filepath, "wb") as f:
        f.write(rep.content)
    z+=1

  i+=step
```
